```python
"""
Cette fonction prend en paramètre une ligne de casting appartement au dataset sachant qu'ils peuvent être parcourus
en itérant sur df["cast"] et retourne une liste [(nom,prenom),(nom,prenom)...etc]
"""
def get_actors_names(cast):
    name = str(cast)
    names = name.split(",")
    l = []
    for i in names:
        n = i.split(" ")
        nn = []
        for i in n:
            if(i!=""):
                nn.append(i)
        l.append(nn)
    actors= []
    for i in l:
        actors.append( tuple([i[-1]," ".join(i[:-1])]))
    return actors
```

```python
"""
Cette fonctions sert a ajouter le contenu d'une column automatiquement en donnant le nom de la colonne et celui
du concept correspondant a la colonne.
Pas besoin de la coder, regardez comment elle fonctionn et servez vous en.
Remarques : 
    - np.unique(L) renvoie une liste d'éléments unique d'un tableau numpy.
    
"""
def create_concept_from_col(df,column,concept_name):
    graphe = rdflib.Graph()
    concept_t = rdflib.URIRef(myont_ns+concept_name)
    graphe.add((concept_t,rdflib.RDF.term("type"),rdflib.OWL.term("Class")))
    arr = np.array(df[column]).astype(str)
    s = set()
    for i in np.unique(arr):
        L = i.replace(" ","").replace('.',"").replace('"',"").split(",")
        s.update(set(L))
    for i in s:
        graphe.add((rdflib.URIRef(rdf_url+i),rdflib.RDF.term("type"),concept_t))
    return graphe
```
