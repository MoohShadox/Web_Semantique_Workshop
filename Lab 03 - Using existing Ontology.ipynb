```python
import requests
import json

query = "select distinct ?Concept where {[] a ?Concept} LIMIT 100"
url = f"""https://dbpedia.org/sparql?default-graph-uri=http%3A%2F%2Fdbpedia.org&query={query}&format=application%2Fsparql-results%2Bjson&CXML_redir_for_subjs=121&CXML_redir_for_hrefs=&timeout=30000&run=+Run+Query+"""

# Adding empty header as parameters are being sent in payload
r = requests.post(url)
json_data = json.loads(r.text)
```

```python
for i in dict(json_data):
    print(i)
```

```python
for i in dict(json_data["results"]):
    print(i)
```

```python
for i in json_data["results"]["bindings"]:
    print(i)
```

```python
for i in json_data["results"]["bindings"]:
    print(i["Concept"]["value"])
```
